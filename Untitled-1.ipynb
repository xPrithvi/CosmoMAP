```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import gc

# ✅ Step 1: Use `np.memmap` to load `.npy` files without RAM overload
print("Mapping datasets to memory...")
X_mass = np.load("/home/amanbh/projects/tf217/DISS/Maps_Mcdm_IllustrisTNG_LH_z=0.00.npy", mmap_mode='r')
X_vel  = np.load("/home/amanbh/projects/tf217/DISS/Maps_Vcdm_IllustrisTNG_LH_z=0.00.npy", mmap_mode='r')
Y_gas  = np.load("/home/amanbh/projects/tf217/DISS/Maps_Mgas_IllustrisTNG_LH_z=0.00.npy", mmap_mode='r')
Y_temp = np.load("/home/amanbh/projects/tf217/DISS/Maps_T_IllustrisTNG_LH_z=0.00.npy", mmap_mode='r')

print(f"Total samples: {len(X_mass)}")  # Check dataset size

# ✅ Step 2: Efficient Normalization (applied dynamically per batch)
def normalize_batch(batch, epsilon=1e-6):
    min_val = np.min(batch)
    max_val = np.max(batch)
    return np.log1p(batch - min_val + epsilon) / np.log1p(max_val - min_val + epsilon)

# ✅ Step 3: Create a Memory-Efficient Data Generator
def data_generator():
    for i in range(len(X_mass)):  # Read one sample at a time
        X1 = normalize_batch(X_mass[i])  # Dark matter mass
        X2 = normalize_batch(X_vel[i])   # Dark matter velocity
        Y1 = normalize_batch(Y_gas[i])   # Gas density
        Y2 = normalize_batch(Y_temp[i])  # Temperature
        yield (np.stack([X1, X2], axis=-1), np.stack([Y1, Y2], axis=-1))

# ✅ Step 4: Convert Generator to TensorFlow Dataset
dataset = tf.data.Dataset.from_generator(
    data_generator, 
    output_signature=(
        tf.TensorSpec(shape=(256, 256, 2), dtype=tf.float32),  # Change shape if needed
        tf.TensorSpec(shape=(256, 256, 2), dtype=tf.float32)
    )
)

# ✅ Step 5: Preprocess the Data Pipeline
dataset = dataset.shuffle(1000).batch(8).prefetch(tf.data.AUTOTUNE)  # Efficient loading

# ✅ Step 6: Define U-Net Model (or any model you want)
def unet_model(input_shape):
    inputs = tf.keras.layers.Input(input_shape)

    # Encoder
    conv1 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    conv1 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(conv1)
    pool1 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(conv1)

    conv2 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(pool1)
    conv2 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(conv2)
    pool2 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(conv2)

    conv3 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(pool2)
    conv3 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(conv3)
    pool3 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(conv3)

    # Bottleneck
    conv4 = tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same')(pool3)
    conv4 = tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same')(conv4)

    # Decoder
    up1 = tf.keras.layers.UpSampling2D(size=(2, 2))(conv4)
    up1 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(up1)
    merge1 = tf.keras.layers.Concatenate()([up1, conv3])

    up2 = tf.keras.layers.UpSampling2D(size=(2, 2))(merge1)
    up2 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(up2)
    merge2 = tf.keras.layers.Concatenate()([up2, conv2])

    up3 = tf.keras.layers.UpSampling2D(size=(2, 2))(merge2)
    up3 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(up3)
    merge3 = tf.keras.layers.Concatenate()([up3, conv1])

    outputs = tf.keras.layers.Conv2D(2, (1, 1), activation='sigmoid')(merge3)

    model = tf.keras.models.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# ✅ Step 7: Initialize and Train Model on Efficient Data Pipeline
model = unet_model((256, 256, 2))  # Change input shape as needed
model.summary()

# 🚀 Train the model with streaming data
model.fit(dataset, epochs=10)
```

```python
# Calculate power spectra of predictions vs. ground truth
from scipy import fftpack

def power_spectrum(img):
    # Convert TensorFlow tensor to NumPy array
    if tf.is_tensor(img):
        img = img.numpy()
    # Ensure it's a float array
    img = np.asarray(img, dtype=np.float32)
    # Calculate FFT
    img_fft = fftpack.fft2(img)
    img_power = np.abs(fftpack.fftshift(img_fft))**2
    return img_power

# Make sure test_batch is a NumPy array
sample_idx = 0
# Gas density power spectrum
true_power = power_spectrum(test_batch[1][sample_idx,:,:,0])
pred_power = power_spectrum(predictions[sample_idx,:,:,0])
# Plot and compare
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.imshow(np.log10(true_power), cmap='viridis')
plt.title('Ground Truth Power Spectrum')
plt.subplot(122)
plt.imshow(np.log10(pred_power), cmap='viridis')
plt.title('Predicted Power Spectrum')
plt.tight_layout()
plt.show()
```

```python
from skimage.metrics import structural_similarity as ssim
import numpy as np
import tensorflow as tf

def calculate_metrics(y_true, y_pred):
    # Flatten for overall metrics
    y_true_flat = y_true.reshape(-1)
    y_pred_flat = y_pred.reshape(-1)
    
    # Basic metrics
    mae = np.mean(np.abs(y_true_flat - y_pred_flat))
    mse = np.mean(np.square(y_true_flat - y_pred_flat))
    rmse = np.sqrt(mse)
    
    # Calculate PSNR - clip values to avoid infinite
    max_val = np.max(y_true)
    psnr = 20 * np.log10(max_val / np.sqrt(mse)) if mse > 0 else 100.0
    
    # Calculate SSIM for each channel
    ssim_values = []
    for c in range(y_true.shape[-1]):
        # Check image dimensions and use smaller window size if needed
        min_dim = min(y_true.shape[0], y_true.shape[1])
        win_size = min(7, min_dim - (min_dim % 2) + 1)  # Ensure it's odd and fits
        
        # Only calculate SSIM if possible
        if min_dim >= 5:  # Minimum window size of 5 for SSIM
            ssim_val = ssim(y_true[..., c], y_pred[..., c], 
                           data_range=np.max(y_true[..., c]) - np.min(y_true[..., c]),
                           win_size=win_size)
            ssim_values.append(ssim_val)
    
    avg_ssim = np.mean(ssim_values) if ssim_values else 0.0
    
    return {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'PSNR': psnr,
        'SSIM': avg_ssim
    }

# Create a test dataset from validation data (use a smaller number of batches for testing)
test_dataset = dataset.take(5)  # Take 5 batches for evaluation

# Dictionary to store metrics
all_metrics = []

# Get predictions and calculate metrics
for inputs, ground_truth in test_dataset:
    predictions = model.predict(inputs)
    
    # Add a check to ensure inputs and outputs match in shape
    if predictions.shape != ground_truth.numpy().shape:
        print(f"Warning: Shape mismatch - Ground truth: {ground_truth.shape}, Predictions: {predictions.shape}")
        continue
        
    batch_metrics = calculate_metrics(ground_truth.numpy(), predictions)
    all_metrics.append(batch_metrics)
    
    # Print individual batch metrics
    print(f"Batch metrics: MAE={batch_metrics['MAE']:.4f}, MSE={batch_metrics['MSE']:.4f}, PSNR={batch_metrics['PSNR']:.2f}")

# Calculate average metrics across all batches
if all_metrics:
    avg_metrics = {metric: np.mean([batch[metric] for batch in all_metrics]) 
                  for metric in all_metrics[0].keys()}

    print("\nModel Evaluation Metrics (Average):")
    for metric, value in avg_metrics.items():
        print(f"{metric}: {value:.6f}")
else:
    print("No metrics were calculated successfully.")
```
